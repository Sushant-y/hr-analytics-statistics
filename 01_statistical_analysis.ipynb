# HR Analytics - Statistical Analysis

Purpose
This notebook presents statistics and pobability focused analysis of HR analytics dataset.

The objective is to determine correct statistical reasoning, appropriate test selection,
explicit assumprtions checks, and correct interpretation of results.

This is an observational dataset no casual claimns are made.

## Dataset Context

The dataset contains employee-level HR information collected through routine
organizational processes.

Important implications:
- The data is observational, not experimental
- There is no random assignment or intervention
- Statistical results reflect associations only

As a result, techniques such as A/B testing and causal inference are not applicable.

## Variable Classification Summary

Before analysis, all variables were classified as:
- Binary categorical
- Nominal categorical
- Ordinal categorical
- Continuous numerical

Identifiers and constant-value variables were excluded from analysis.

 All statistical tests in this notebook are chosen based on these classifications.


## Statistical Questions

The following statistical questions guide the analysis.  
Each question is framed to allow formal hypothesis testing based on variable type
and distributional assumptions.

1. Is employee attrition independent of overtime status?

2. Is employee attrition independent of job role?

3. Does monthly income differ across job levels?

4. Is work-life balance associated with employee attrition?

5. Is total working experience correlated with monthly income?
